```python
%load_ext autoreload
%autoreload 2

import pandas as pd
import joblib

from fgclustering.forest_guided_clustering import *
```

```python
data_boston = pd.read_csv('./data/data_boston.csv')
data_boston['CHAS'] = data_boston['CHAS'].astype('category')
rf = joblib.load(open('./data/random_forest_boston.joblib', 'rb'))

optimal_k = forest_guided_clustering(output='forest_guided_clustering_boston', data=data_boston, target_column='target', model=rf,  
                                     max_K = 8, number_of_clusters = 6, max_iter_clustering = 1000, 
                                     bootstraps_JI = 300, discart_value_JI = 0.6, 
                                     bootstraps_p_value = 1000, thr_pvalue = 0.001, random_state = 42)
```
